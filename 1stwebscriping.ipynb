```python
url='https://www.bbc.com/sport/football/premier-league/top-scorers#TopScorers'
```

```python
players_names=[]
Serial_number=[]
team_names=[]
gola_scored=[]
Assists=[]
Total_Match_played=[]
Total_shots=[]
Gola_conversion_rates=[]
shot_Accuracys=[]
```

```python
from bs4 import BeautifulSoup
import requests
try:
    response=requests.get(url)
    response.raise_for_status()
except Exception as e:

    print(e)

else:
    status_code=response.status_code
    soup=BeautifulSoup(response.content,'html.parser')
    players=soup.find('tbody').find_all('tr',class_='ssrcss-qqhdqi-TableRowBody e1icz100')
    for player in players:
        S_no=player.find('div',class_= 'ssrcss-wjp01q-RankAndBadgeWrapper e1n6h4mf2').get_text(strip=True)
        player_name=player.find('div',class_='ssrcss-m6ah29-PlayerName e1n8xy5b1').get_text(strip=True)
        team_name=player.find('div',class_='ssrcss-qvpga1-TeamsSummary e1n8xy5b0').get_text(strip=True)
        gola=int(player.find('td','ssrcss-126tggj-StyledTableData ef9ipf1').get_text(strip=True))
        Recode=player.find_all('div',class_='ssrcss-1vo7v3r-CellWrapper ef9ipf0')
        assist=int(Recode[1].get_text(strip=True))
        match_played=int(Recode[2].get_text(strip=True))
        Total_shot=int(Recode[-3].get_text(strip=True))
        Gola_conversion_rate=Recode[-2].get_text(strip=True)
        Shot_Accuracy=Recode[-1].get_text(strip=True)

        
        players_names.append(player_name)
        Serial_number.append(S_no)
        team_names.append(team_name)
        gola_scored.append(gola)
        Assists.append(assist)
        Total_Match_played.append(match_played)
        Total_shots.append(Total_shot) 
        Gola_conversion_rates.append(Gola_conversion_rate)
        shot_Accuracys.append(Shot_Accuracy)
   
  
```

```python
data={
    'Serial Number':Serial_number,
    'Player Name':players_names,
    'Team Name':team_names,
    'Gola Scored':gola_scored,
    'Assists':Assists,
    'Total Match Played':Total_Match_played,
    'Total Shots':Total_shots,
    'Gola Conversion Rate':Gola_conversion_rates,
    'Shot Accuracy':shot_Accuracys
}
```

```python
import pandas as pd
df=pd.DataFrame(data)
```
